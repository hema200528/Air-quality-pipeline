```python
import pandas as pd
from pathlib import Path

INPUT_FILE = "team_4.parquet"
df = pd.read_parquet(INPUT_FILE)

print(f"Loaded {len(df):,} rows  x  {df.shape[1]} columns")
df.head()
```

```python
print("Columns:", list(df.columns))
print()
print(df.dtypes)
```

```python
print("=" * 60)
print("INGESTION SUMMARY")
print("=" * 60)

print(f"Total records      : {len(df):,}")
print(f"Columns            : {df.shape[1]}")
print(f"Date range         : {df['datetime'].min()}  →  {df['datetime'].max()}")
print(f"Years              : {sorted(df['year'].unique().tolist())}")
print(f"Months             : {sorted(df['month'].unique().tolist())}")
print(f"States             : {df['state'].nunique()}  → {list(df['state'].unique())}")
print(f"Cities             : {df['city'].nunique()}  → {list(df['city'].unique())}")
print(f"Stations           : {df['station_id'].nunique()}")
print(f"Pollutant species  : {df['pollutant'].nunique()}")
print(f"   {sorted(df['pollutant'].unique().tolist())}")
```

```python
summary_lines = [
    f"Total records      : {len(df):,}",
    f"Columns            : {df.shape[1]}",
    f"Date range         : {df['datetime'].min()}  →  {df['datetime'].max()}",
    f"Years              : {sorted(df['year'].unique().tolist())}",
    f"Stations           : {df['station_id'].nunique()}",
    f"Pollutant species  : {df['pollutant'].nunique()}",
    "",
    "Records per year-month:",
    df.groupby(["year", "month"]).size().to_string(),
    "",
    "Records per pollutant:",
    df["pollutant"].value_counts().to_string(),
]
Path("ingestion_summary.txt").write_text("\n".join(summary_lines))
print("✓ Saved ingestion_summary.txt")
```

```python
OUTPUT_DIR = Path("partitioned_data")
OUTPUT_DIR.mkdir(exist_ok=True)

written = 0
for (yr, mo), chunk in df.groupby(["year", "month"]):
    folder = OUTPUT_DIR / f"year={yr}" / f"month={mo:02d}"
    folder.mkdir(parents=True, exist_ok=True)
    out_path = folder / "data.parquet"
    chunk.to_parquet(out_path, index=False)
    written += 1
    print(f"  year={yr} month={mo:02d}  →  {len(chunk):>8,} rows")

print(f"\n✓ Wrote {written} partition files into '{OUTPUT_DIR}/'")
```

```python
import shutil
OUTPUT_DIR = Path("partitioned_data")

# wipe the old broken output
if OUTPUT_DIR.exists():
    shutil.rmtree(OUTPUT_DIR)
OUTPUT_DIR.mkdir()

written = 0
for (yr, mo), chunk in df.groupby(["year", "month"]):
    folder = OUTPUT_DIR / f"year={yr}" / f"month={mo:02d}"
    folder.mkdir(parents=True, exist_ok=True)
    # drop year/month from the data — they're encoded in the folder path
    chunk.drop(columns=["year", "month"]).to_parquet(folder / "data.parquet", index=False)
    written += 1
    print(f"  year={yr} month={mo:02d}  →  {len(chunk):>8,} rows")

print(f"\n✓ Wrote {written} partition files into '{OUTPUT_DIR}/'")
```
